```python
data1 = data[data["Year"] < 2012].copy()
data2 = data[data["Year"] >= 2012].copy()
xtrain = data1[["BMI", "Schooling", "Income composition of resources", "Life expectancy"]]
xtest = data2[["BMI","Schooling", "Income composition of resources"]]
```

```python

mean = xtrain.mean()
std = xtrain.std()

thres = 3

outliers = []
for col in xtrain.columns:
    z_scores = (xtrain[col] - mean[col]) / std[col]
    outliers.extend(xtrain[np.abs(z_scores) > thres].index.tolist())
    
outliers = list(set(outliers))

xtrain_cleaned = xtrain.drop(index=outliers)

print("Identified outliers:", outliers)
print("Cleaned dataframe:")
print(xtrain_cleaned)
ytrain = xtrain["Life expectancy"]
xtrain = xtrain.drop(columns=["Life expectancy"])
ytest = data2["Life expectancy"]
```

```python
class Linear_Regression():
    def __init__(self, lr=0.001, n_iters=100000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None
    
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        for _ in range(self.n_iters):
            y_pred = np.dot(X, self.weights) + self.bias
            dw = (1/n_samples) * np.dot(X.T, (y_pred - y))
            db = (1/n_samples) * np.sum(y_pred - y)
            
            self.weights -= self.lr * dw
            self.bias -= self.lr * db
    
    def predict(self, X):
        print(X.shape)
        y_pred = np.dot(X, self.weights) + self.bias
        return y_pred
```

```python

def train_model(xtrain, ytrain, xtest, ytest, model, eta):
    model.fit(xtrain, ytrain)
    y_pred_train = model.predict(xtrain)
    y_pred_test = model.predict(xtest)
    
    train_err = np.mean((y_pred_train - ytrain) ** 2)
    test_err = np.mean((y_pred_test - ytest) ** 2)
    return model.weights, train_err, test_err
```

```python
# xtrain = xtrain.drop(columns = ["BMI"])
# xtest = xtest.drop(columns = ["BMI"])
d = xtrain.shape[1]
print(d)
w = np.zeros((d, 1))
eta = 0.001
model = Linear_Regression(lr=eta)
w, train_err, test_err = train_model(xtrain, ytrain, xtest, ytest, model, eta)
print(w)
print(train_err)
print(test_err)
```

```python
X = xtrain
y = ytrain
X_b = np.c_[np.ones((X.shape[0], 1)), X]  # Add bias term
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
bias = theta_best[0]
weights = theta_best[1:]
print(weights)
y_pred_test = np.dot(xtest, weights) + bias
test_err = np.mean((y_pred_test - ytest) ** 2)
print(test_err)
```
