**Kernel PCA**

 Kernel PCA is a non-linear dimensionality reduction technique that extends PCA. It uses the kernel trick to project the data into a higher-dimensional space and then applies PCA in that space. Read the theory for kernel PCA [here](https://people.eecs.berkeley.edu/~wainwrig/stat241b/scholkopf_kernel.pdf).


**Kernel PCA for Radial Basis Function(RBF)**
 
 Go through the [given link](https://rasbt.github.io/mlxtend/user_guide/feature_extraction/RBFKernelPCA/). Try to implement the algorithm given here using numpy on the same example used in the link.



```python
from sklearn.datasets import make_circles

X, y = make_circles(n_samples=1000, random_state=123, 
                    noise=0.1, factor=0.2)

plt.figure(figsize=(8,6))

plt.scatter(X[y==0, 0], X[y==0, 1], color='red', alpha=0.5)
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', alpha=0.5)
plt.title('Concentric circles')
plt.ylabel('y coordinate')
plt.xlabel('x coordinate')
plt.show()
```

You may also want to apply PCA on the above data and understand the difference between PCA and Kernel PCA. Essentially, Kernel PCA is used when the data is non-linearly separable.
